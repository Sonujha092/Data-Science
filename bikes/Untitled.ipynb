```python
def plot_by_hour(data, year=None):
    df = data
    if year: df = df[df.datetime.dt.year == year]
    df.loc[:, ('hour')] = df.datetime.dt.hour
    by_hour = df.groupby(['hour', 'year'])['count'].sum().unstack()
    return by_hour.plot.bar(figsize=(15,5))

plot_by_hour(df)
```

```python
def plot_by_year(data, att):
    df = data
    df['year'] = df.datetime.dt.year
    df['month'] = df.datetime.dt.month
    df['hour'] = df.datetime.dt.hour
    by_month = df.groupby([att, 'year']).sum()['count'].unstack()
    return by_month.plot.bar(figsize=(15,5))
        
plot_by_year(df, 'month')
plot_by_year(df,'hour')
```
